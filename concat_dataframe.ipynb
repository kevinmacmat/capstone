```python
batch_df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12, df_13], ignore_index = True)
```

```python
# Add comments to sqr dataframe
df['comments'] = batch_df['comments']
```

```python
# Remove schools with no comments
df = df[df['comments'] != '[]']
```

```python
# Reset index
df = df.reset_index(drop=True)
```

```python
# Create borough column
df['borough'] = df.dbn.str.extract(r'([A-Z]+)',expand=False)
```

```python
# Change dbn letter to borough names
df['borough'] = df['borough'].replace('M', 'manhattan')
df['borough'] = df['borough'].replace('K', 'brooklyn')
df['borough'] = df['borough'].replace('R', 'staten_island')
df['borough'] = df['borough'].replace('X', 'bronx')
df['borough'] = df['borough'].replace('Q', 'queens')
```

```python
# Add sqr_score as an average of fam_comm_ties_rating and trust_rating. 
df['sqr_rating'] = df.loc[:,['fam_comm_ties_rating','trust_rating']].mean(axis=1)
```

If we take the average of the fam_comm_ties_rating and trust_rating scores we can get one of seven values: .250, .375, .500, .625, .750, .875, 1.00
